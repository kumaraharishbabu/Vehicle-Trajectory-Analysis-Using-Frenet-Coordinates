```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all sheets from the Excel file
file_path = r'C:\Users\kumar\OneDrive\Frenet_Output_TW\all_sheets_csvs\all_frenet_combined.xlsx'
sheets_dict = pd.read_excel(file_path, sheet_name=None)

# Plot
plt.figure(figsize=(12, 6))

for sheet_name, sheet_df in sheets_dict.items():
    if 's_arc_length' in sheet_df.columns and 'd_lateral_offset' in sheet_df.columns:
        plt.plot(sheet_df['s_arc_length'], sheet_df['d_lateral_offset'], '.', markersize=1, alpha=0.5)

plt.xlabel('Longitudinal Distance (s) [arc length]')
plt.ylabel('Lateral Offset (d)')
plt.title('s-d Trajectories for TW Vehicles')
plt.grid(True)
plt.tight_layout()
plt.show()
```
